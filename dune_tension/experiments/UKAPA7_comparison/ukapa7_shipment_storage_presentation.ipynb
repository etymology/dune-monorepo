## Generate the plots

This notebook calls the same plotting functions used by the Quarto slide deck, writes the PNG assets into the experiment directory, and then displays those generated files in the presentation cells below.

```python
from pathlib import Path
import sys

import matplotlib.image as mpimg
import matplotlib.pyplot as plt


def find_repo_root(start: Path) -> Path:
    for path in (start, *start.parents):
        if (path / "pyproject.toml").exists() and (path / "src/dune_tension").exists():
            return path
    raise RuntimeError("Could not find dune-monorepo root")


ROOT = find_repo_root(Path.cwd())
SRC = ROOT / "src"
if str(SRC) not in sys.path:
    sys.path.insert(0, str(SRC))

from dune_tension.ukapa7_comparison.generate_landscape_display_plots import (
    save_layer_change_in_tension_plot,
    save_layer_landscape_plot,
)

EXPERIMENT_DIR = ROOT / "dune_tension/experiments/UKAPA7_comparison"
G_ACTION_JSON = ROOT / "dune_tension/apa_uk7g.json"
U_ACTION_JSON = ROOT / "dune_tension/UKAPA7U.json"
G_SUMMARY_CSV = ROOT / "dune_tension/data/tension_summaries/tension_summary_UKAPA7_G.csv"
U_SUMMARY_CSV = ROOT / "dune_tension/data/tension_summaries/tension_summary_UKAPA7_U.csv"

plots = {
    "g_landscape": EXPERIMENT_DIR / "ukapa7_landscape_G.png",
    "g_change": EXPERIMENT_DIR / "ukapa7_change_in_tension_G.png",
    "u_landscape": EXPERIMENT_DIR / "ukapa7_landscape_U.png",
    "u_change": EXPERIMENT_DIR / "ukapa7_change_in_tension_U.png",
    "g_profile": ROOT / "dune_tension/data/tension_plots/tension_profile_cloud_G_mode_allsamples_cov0p5_it3_bins40.png",
    "u_profile": ROOT / "dune_tension/data/tension_plots/tension_profile_cloud_U_noscale_allsamples_cov0p5_it3_bins40.png",
}


def regenerate_plots() -> None:
    save_layer_landscape_plot(
        action_json=G_ACTION_JSON,
        summary_csv=G_SUMMARY_CSV,
        layer="G",
        output_path=plots["g_landscape"],
    )
    save_layer_change_in_tension_plot(
        action_json=G_ACTION_JSON,
        summary_csv=G_SUMMARY_CSV,
        layer="G",
        output_path=plots["g_change"],
    )
    save_layer_landscape_plot(
        action_json=U_ACTION_JSON,
        summary_csv=U_SUMMARY_CSV,
        layer="U",
        output_path=plots["u_landscape"],
    )
    save_layer_change_in_tension_plot(
        action_json=U_ACTION_JSON,
        summary_csv=U_SUMMARY_CSV,
        layer="U",
        output_path=plots["u_change"],
    )


def show_image(path: Path, *, figsize: tuple[float, float] = (14, 8)) -> None:
    image = mpimg.imread(path)
    fig, ax = plt.subplots(figsize=figsize)
    ax.imshow(image)
    ax.axis("off")
    plt.show()


regenerate_plots()
plots
```

## Measurement conditions

- In the absence of a DWA we were able to measure the outer two layers of wires before disassembly.
- **G layer:** full comparison on both sides, `481` aligned wires on side A and `481` on side B.
- **U layer:** partial Chicago coverage only, with `385` aligned wires on side A and `641` on side B.
- U layer wires were accessed through a "slit" cut in the G wires near the middle of the APA.
- The measurements were taken in similar but not identical ways
- The frame is in a similar state for the G layer but not for the U layer due to the added tension of the G layer over it.

## G Layer

- Most wires on both sides changed by `-0.45 N` (Chicago lower).
- On a wire-by-wire basis, differences range from `-1.63` to `+1.32` with `sigma = 0.4`
- Actual increase in tension is less likely than compounded measurement uncertainties

- Most wires dropped about `-0.5N`

## Discussion

- ~0.5N decrease over 2.25y consistent with wire relaxation under stress
- The modal decrease is clear but on the individual wire level, noisy
- Huge drops (`-3N`) probably the result of mistaken measurements at Daresbury

## Recommendations

- Changes in tension even large ones shouldn't disqualify acceptance if the final tension is acceptable.
- Consortium should establish standards for the final acceptable tensions before installation, taking into account decreasing rate of relaxation
- Plausibly, 5.5->5.0 N over 3 years => 5.0->4.8-4.9 N over the next several years
